```python
import numpy as np
import gensim

from src.utils.logger import logger
```

```python
sequences = np.load('../data/books_sequences.npy')
```

```python
sequences.shape
```

```python
seq_samp = sequences[:1000, :]
```

```python
np.save('../data/books_sequences_sample.npy', seq_samp)
```

```python
seq_samp = seq_samp.astype(str).tolist()
```

```python
seq_samp
```

```python
model = gensim.models.Word2Vec(seq_samp.tolist(), size=128, window=5, min_count=1, workers=4, negative=5, seed=42)
```

```python
import logging

logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
```

```python
model.train(seq_samp.tolist(), total_examples=len(seq_samp), epochs=3)
```

```python
model.wv.similarity('0', '99')
```
